```python
#import libraries 
import numpy as np
import pandas as pd
import plotly as py
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('covid_19_india.csv',index_col=0)
df.drop('Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Date'] =   pd.to_datetime(df['Date'], dayfirst=True)
df.head()
```

```python
cvd = df.groupby(['State/UnionTerritory', 'Date']).sum()
cvd = cvd.sort_values('Date', ascending = False).reset_index()
cvd
```

```python
cvd = cvd.drop_duplicates(subset = ['State/UnionTerritory'])
cvd = cvd[cvd['Confirmed']>0]
cvd
```

```python
import folium
state_geo = r'india_states.json'
```

```python
map = folium.Map(location=[21, 78], zoom_start=5)
map
```

```python
map.choropleth(geo_data=state_geo, data=cvd,
         columns=['State/UnionTerritory', 'Confirmed'],
         key_on='feature.id',
         fill_color='YlOrRd', fill_opacity=0.7, line_opacity=0.2,
         legend_name='Coronovirus cases as of 3 July, 2020')
folium.LayerControl().add_to(map)
map
```
